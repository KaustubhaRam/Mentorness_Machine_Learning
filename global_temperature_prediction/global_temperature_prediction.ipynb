```python
def fill_nan_bottom_up(df, column):
    if column not in df.columns:
        print(f"Column '{column}' not found in DataFrame.")
        return df

    for i in range(len(df)):
        if pd.isnull(df.loc[i, column]):
            # Get the next 5 non-null values from the current row forward
            valid_values = df[column].iloc[i+1:].dropna().head(5)
            if len(valid_values) > 0:
                # Replace NaN with the mean of the next 5 values (bottom-up)
                df.loc[i, column] = valid_values.mean()
    return df
```

```python
cols = [
    "LandAverageTemperature", 
    "LandAverageTemperatureUncertainty", 
    "LandMaxTemperature", 
    "LandMaxTemperatureUncertainty", 
    "LandMinTemperature", 
    "LandMinTemperatureUncertainty", 
    "LandAndOceanAverageTemperature", 
    "LandAndOceanAverageTemperatureUncertainty"
]

# Debugging step: Print the actual columns in your DataFrame
print("Available columns in glt:", glt.columns)

# Loop through the columns and apply the function only if the column exists
for col in cols:
    if col in glt.columns:  # Check if column exists before applying function
        glt = fill_nan_bottom_up(glt, col)
    else:
        print(f"Column {col} not found in DataFrame")
```
